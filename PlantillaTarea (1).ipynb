## Modelo Black and Scholes y Valor presente


```python
# Primero calculamos la volatilidad de la accion de TSLA
import pandas_datareader.data as web
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt

start = dt.datetime(2010,1,1)    
end =dt.datetime(2022,10,25) 
symbol = 'TSLA'
source = 'yahoo'
data = web.DataReader(symbol, source, start, end)
data['change'] = data['Adj Close'].pct_change()
data['rolling_sigma'] = data['change'].rolling(20).std() * np.sqrt(255)


data.rolling_sigma.plot()
plt.ylabel('$\sigma$')
plt.title('TSLA Rolling Volatility')
plt.legend()
plt.show()
```

Este modelo no corre en Google Collab, pero si en Visual Studio Code, y nos arroja la volatilidad de la accion en los ultimos años y la pronosticada de ahora tal como se puede apreciar a continuacion: Figure_1.png

Gracias a esto calculamos que la volatilidad para utilizar en el modelo Black and Scholes es de un aproximado 0.696.

```python
#Modelo Black and Scholes y VP
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt


def geo_paths(S, T, r, q, sigma, steps, N):
    
    dt = T/steps
    #S_{T} = ln(S_{0})+\int_{0}^T(\mu-\frac{\sigma^2}{2})dt+\int_{0}^T \sigma dW(t)
    ST = np.log(S) +  np.cumsum(((r - q - sigma**2/2)*dt +\
                              sigma*np.sqrt(dt) * \
                              np.random.normal(size=(steps,N))),axis=0)
    
    return np.exp(ST)

S_TSLA = 222.41
K_TSLA = 212.5
r_TSLA = 0.0325
T_TSLA = 1/12
sigma_TSLA = 0.696

S = S_TSLA #stock price S_{0}
K = K_TSLA # strike
T = T_TSLA # time to maturity
r = r_TSLA # risk free risk in annual %
q = 0.0 # annual dividend rate
sigma = sigma_TSLA # annual volatility in %
steps = 5 # time steps
N = 100000 # number of trials

paths= geo_paths(S,T,r, q,sigma,steps,N)


def black_scholes_call(S,K,T,r,q,sigma):
    d1 = (np.log(S/K) + (r - q + sigma**2/2)*T) / sigma*np.sqrt(T)
    d2 = d1 - sigma* np.sqrt(T)
    
    call = S * np.exp(-q*T)* norm.cdf(d1) - K * np.exp(-r*T)*norm.cdf(d2)
    return call



payoffs = np.maximum(paths[-1]-K, 0)
option_price = np.mean(payoffs)*np.exp(-r*T) #discounting back to present value

bs_price = black_scholes_call(S,K,T,r,q,sigma)

print(f"Black Scholes Price is {bs_price}")
print(f"Simulated price is {option_price}")



names = ['BS_Model', 'VP_Model']
values = [bs_price,option_price]
plt.scatter(names, values)
plt.suptitle('Option Prices')
plt.show()
```

En el modelo Black and Scholes, tal como se aprecia en el codigo, nuestors coeficientes son:

Precio accion "S" = 222.41 (En la fecha de 26/10/2022)

Precio Strike "K" = 212.5 

Tiempo maduracion "T" = 1 mes, debido a que esa es la fecha limite del precio Strike anterior.

Tasa libre de riesgo "r" = 0.0325, la cual es la tasa libre de riesgo de los estados unidos actualmente.

Volatilidad "sigma" = 0.696, la cual es la calculada con el codigo anterior.

El precio del modelo de Black and Scholes nos arroja 22.276 y el precio del modelo el cual trae a valor presente conjunto a una simulacion de monte carlos en un periodo de 5 años y con repeticion de 100mil veces es de 23.143. 

Esta diferencia se puede acortar aumentando el numero de repeticiones de Monte Carlo, debido al teorema del limite central, el cual nos dice que la distribucion se acercara a una distribucion normal en cuanto mas crezca el numero de repeticiones que utilicemos en el modelo.
